# Primary factors which determine the presence/absence of each of the 3 species

In this project, we use the data obtained from cameras at 19 different sites in the Serengeti Park in Tanzania to answer the question "What are the primary factors which determine the presence/absence of each of the 3 species"

# Summary of the project

$\textit{What are the primary factors which determine the presence/absence of each of the 3 species?}$

Topi

$amRivDist$ and $LriskDry$ are the primary factors that determine the presence/absence of topi. Specifically, with all other things unchanged, every unit increase in $amRivDist$ will decrease the odds of topi presence by $e^{-0.0003}=0.9997$ times, and every 0.001 unit(since $LriskDry$ value is at 0.001 level) increase in $LriskDry$ will increase the odds of topi presence by $e^{1.61}=5.05$ times.

$$
\begin{aligned}
log\frac{\hat{\mu}_{topi}}{1-\hat{\mu}_{topi}} & = \hat{\beta_0}+\hat{\beta_1}*x_{amRivDist} + \hat{\beta_2}*x_{LriskDry}\\
& =  -2.5573-0.0003*x_{amRivDist}  +161.9892 *x_{LriskDry}
\end{aligned}
$$

Zebra

$TM100$, $T50$ and $month$ are the primary factors that determine the presence/absence of zebra. Specifically, with all other things unchanged, every unit increase in $TM100$ will decrease the odds of topi presence by $e^{-0.2718}=0.762$ times, and every unit increase in $T50$ will increase the odds of topi presence by $e^{0.001}=1.001$ times, and every unit increase in $month$ will increase the odds of topi presence by $e^{0.0505}=1.05$ times. One thing to notice here is that Zebra will show up more as in later of a year and that's because Tanzania is in the south part of the earth.

$$
\begin{aligned}
log\frac{\hat{\mu}_{zebra}}{1-\hat{\mu}_{zebra}} & = \hat{\beta_0}+\hat{\beta_1}*x_{TM100} + \hat{\beta_2}*x_{T50} +  \hat{\beta_3}*x_{month}\\
& =  -0.8747-0.2718*x_{TM100}  +0.0010 *x_{T50}+ 0.0505x_{month}
\end{aligned}
$$

Gazelle

$ndvi$, $amRivDist$ and $TM100$ are the primary factors that determine the presence/absence of gazelle. Specifically, with all other things unchanged, every unit increase in $ndvi$ will decrease the odds of gazelle presence by $e^{-5.5086}=0.004$ times, and every unit increase in $amRivDist$ will increase the odds of gazelle presence by $e^{0.0002}=1.0002$ times, and every unit increase in $TM100$ will increase the odds of gazelle presence by $e^{-0.7992}=0.449$ times. Notice that for gazelle the glm doesn't have an intercept.

$$
\begin{aligned}
log\frac{\hat{\mu}_{gazelle}}{1-\hat{\mu}_{gazelle}} & = \hat{\beta_1}*x_{ndvi} +\hat{\beta_2}*x_{amRivDist} + \hat{\beta_3}*x_{TM100}\\
& = -5.5086 *x_{ndvi} +0.0002*x_{amRivDist}  -0.7992*x_{TM100}
\end{aligned}
$$

$\textit{A biologist might argue that two neighboring time steps are not independent, since if a species were present at a location it may be more} \textit{ likely to be present at the same location 8 days later.}$

The glms in 'Is previous species presence influence current species presence' section shows that previous presence is significant in modeling current presence. But we cannot conclude that $present$ variable is not indpendent with each other cause topi shows up one week ago will not influence their presence one week later.

## Generalized linear model for presence data  <a class="anchor" id="chapter3"></a>

The interest of this project is the primary factors that determine the  presence or absence of each of the 3 species. Therefore, we can use generalized linear model to forecast presence data to see which variables are significant in the glm model. Then by dropping variables from the glm model we can determine the primary factors that influence the presence or absence of the 3 species.

### Distribution family and link function <a class="anchor" id="section_3_1"></a>

```python
Topi.dtypes
```

Notice all the variables in the cleaned data sets are numerical except siteID. Therefore, an encoding is necessary before we fit a generalized linear model.

```python
Topi['siteID']=Topi['siteID'].astype('category')

# using label encoding for site ID 
Topi['siteID_cat']=Topi['siteID'].cat.codes 
Topi['siteID_cat']=Topi['siteID_cat'] + 1

Topi=Topi.drop('siteID',axis=1)

# rename topi.present to avoid ambugity in formula expression
Topi=Topi.rename(columns={'topi.present':'present'})
Topi.dtypes
```

The presence variable is a binary variable that contains only two values 1 and 0. 1 represents the specy shows up in the past 8 day period. Thus, we will use binomial distribution as the random component and use logit of the binomial mean as the systematic component.

random component: $ ym \sim Bin(\mu,m)$  

systematic component: $log\frac{\mu}{1-\mu} = X^{T}\beta$ 

where $y$ is the $present$ variable, $m$ is the total number of data entries in the data set, and $X$ is the total other variables excluding $present$. 

### Best subset selection for generalized linear model <a class="anchor" id="section_3_2"></a>

The more variables the more information we can get, but it is not true for model selection. Instead of using all the variables we have, we can run a best subset selection that run all the combinations of the variables we have and select the 'best' one based on the metric we choose. Here we use AIC as the metric for best subset selection.

```python
# define function for glm best subset selection
from tqdm import tnrange, tqdm_notebook

def fit_glm(X,Y):
    # fit generalized linear model and return AIC 
    # for this glm we use binomial distribution and logit link function
    # require package statsmodels.api as sm and statsmodels.formula.api as smf
    
    # add intercept manually
    X.insert(loc=0,column='Intercept',value=[1]*len(X))
    model = sm.GLM(Y,X,
                    family=sm.families.Binomial(sm.families.links.logit())).fit()
    return model

def best_subset_glm(X,Y):
    # best subset selection for glm model in this project
    AIC_min = 99999999999999999
    k = len(X.columns)
    
    # loop over k=1 to all features
    for k in tnrange(1,len(X.columns) + 1, desc = 'Loop...'):
        #Looping over all possible combinations: from 11 choose k
        for combo in itertools.combinations(X.columns,k):
            tmp_result = fit_glm(X[list(combo)],Y)   #Store temp result 
            if tmp_result.aic < AIC_min:
                AIC_min = tmp_result.aic
                final_model = tmp_result

    return final_model
```

### Fitting the models <a class="anchor" id="section_3_3"></a>

### Topi presence glm model

```python
X_topi = Topi[['ndvi','fire','amRivDist','TM100',
                    'LriskDry','T50','month', 'siteID_cat']]
Y_topi = Topi['present']
glm_bin_topi = best_subset_glm(X_topi,Y_topi)
print(glm_bin_topi.summary())
```

The above glm model summary shows that amRivDist, TM100, LriskDry and siteID_cat are significant, given significant level of 0.05, which means they are the primary factors that determine the presence/absence of Topi. However, month here is not significant in the glm model which is a contradiction with the above time series analysis result. It might because there exists some nonlinear relationship between month and presence.

Therefore, our full glm model systematic component for Topi presence are
$$
\begin{aligned}
log\frac{\hat{\mu}_{topi}}{1-\hat{\mu}_{topi}} & = \hat{\beta_0} + \hat{\beta_1}*x_{amRivDist} + \hat{\beta_2}*x_{TM100}+\hat{\beta_3}*x_{LriskDry}+
\hat{\beta_4}*x_{siteID\_cat}\\
& = -1.5514 -0.0005*x_{amRivDist} +0.6316*x_{TM100}+131.2553*x_{LriskDry} -0.0702*x_{siteID_cat}
\end{aligned}
$$

where $\hat{\mu}$ is the probability that Topi is present at the given time window. The $\frac{\hat{\mu}}{1-\hat{\mu}}$ is the odds that Topi show up at a given time. Then we can rewrite the above equation to the odds format as following

$$
\begin{align}
\frac{\hat{\mu}_{topi}}{1-\hat{\mu}_{topi}} = e^{-1.5514 -0.0005*x_{amRivDist} +0.6316*x_{TM100}+131.2553*x_{LriskDry} -0.0702*x_{siteID_cat}}
\end{align}
$$

We use the coefficient of $x_{TM100}$ as an example to show how to interpret the coefficients in this glm model, and the interpretation is as following

given all other things unchanged,  one unit increase of $x_{TM100}$ will increase the odds of Topi presence by $e^{0.6285} = 1.87$ times and vice versa.

The sign of the coefficients show how the odds change along with the corresponding variables. The positive coefficients mean increasing variables' value will lead to increasing odds and negative coefficients mean decreasing variables' value will decrease odds. The absolute value of the coefficients reveal the scale of influence that each variable has on the presence of Topi.

The above glm model summary shows that amRivDist, LriskDry, T50, month, and siteID_cat are significant. One interesting thing is that in this glm model with Zebra data set month is significant in the model.

Therefore, our full glm model systematic component for Topi presence are
$$
\begin{aligned}
log\frac{\hat{\mu}_{zebra}}{1-\hat{\mu}_{zebra}} & = \hat{\beta_1}*x_{amRivDist} + \hat{\beta_2}*x_{LriskDry}+\hat{\beta_3}*x_{T50}+\hat{\beta_4}*x_{month} +\hat{\beta_5}*x_{siteID\_cat} \\
& =  -0.0002*x_{amRivDist}  -93.6661*x_{LriskDry}+0.0010*x_{T50}+0.0515*x_{month} -0.0569*x_{siteID\_cat}
\end{aligned}
$$

where $\hat{\mu}$ is the probability that Zebra is present at the given time window. The $\frac{\hat{\mu}}{1-\hat{\mu}}$ is the odds that Topi show up at a given time. Then we can rewrite the above equation to the odds format as following

$odds_{Gazelle show up}$=$\begin{align}
\frac{\hat{\mu}_{zebra}}{1-\hat{\mu}_{zebra}} = e^{-0.0002*x_{amRivDist}  -93.6661*x_{LriskDry}+0.0010*x_{T50}+0.0515*x_{month} -0.0569*x_{siteID\_cat}}
\end{align}$

We did the coefficient interperation in previous section, and it's similar to how we interpret coefficients in this section and next section. The only thing that interested us here is $x_{month}$ variable. $x_{month}$ variable has an coefficient of 0.047 meaning that the later in a year the higher the chance to see a zebra in those site.

$odds_{Gazelle show up}$ =$
\begin{align}
\frac{\hat{\mu}_{gazelle}}{1-\hat{\mu}_{gazelle}} = e^{2.1865 -6.4344*x_{ndvi}-0.1141*x_{siteID\_cat}}
\end{align}
$

# Is previous species presence influence current presence? <a class="anchor" id="chapter5"></a>

A biologist might argue that two neighboring time steps are not independent, since if
a species were present at a location it may be more likely to be present at the same
location 8 days later. Therefore, we need to check how previous presence will effect current presence. 

Notice that previous presence might have correlation with current presence, but it is not saying that presence data are dependent with each other. Therefore, we can still use generalized linear model to test whether previous presence will effect current presence. The followings are the new glm model based on the 'lagging' datasets for three species. We will call these new generalized linear models as 'lagging' models.

```python
# add previous presence column
pre_presence=Topi['present'][:-1].values
Topi_lag =Topi[1:]
Topi_lag['pre_present']=pre_presence

# using best subset selection to fit a generalized linear model
X_topi_lag = Topi_lag[['ndvi','fire','amRivDist','TM100',
                    'LriskDry','T50','month', 'siteID_cat','pre_present']]
Y_topi_lag = Topi_lag['present']
glm_bin_topi_lag = best_subset_glm(X_topi_lag,Y_topi_lag)
print(glm_bin_topi_lag.summary())
```

```python
# add previous presence column
pre_presence_z=Zebra['present'][:-1].values
Zebra_lag =Zebra[1:]
Zebra_lag['pre_present']=pre_presence_z

# using best subset selection to fit a generalized linear model
X_Zebra_lag = Zebra_lag[['ndvi','fire','amRivDist','TM100',
                    'LriskDry','T50','month', 'siteID_cat','pre_present']]
Y_Zebra_lag = Zebra_lag['present']
glm_bin_Zebra_lag = best_subset_glm(X_Zebra_lag,Y_Zebra_lag)
print(glm_bin_Zebra_lag.summary())
```

```python
# add previous presence column
pre_presence_g=Gazelle['present'][:-1].values
Gazelle_lag =Gazelle[1:]
Gazelle_lag['pre_present']=pre_presence_g

# using best subset selection to fit a generalized linear model
X_Gazelle_lag = Gazelle_lag[['ndvi','fire','amRivDist','TM100',
                    'LriskDry','T50','month', 'siteID_cat','pre_present']]
Y_Gazelle_lag = Gazelle_lag['present']
glm_bin_Gazelle_lag = best_subset_glm(X_Gazelle_lag,Y_Gazelle_lag)
print(glm_bin_Gazelle_lag.summary())
```

From the summaies of the three species' model, we find that $pre\_present$ are very significant in all of the models and it plays very positive effect on $present$. Let's check the AIC of each model

```python
print("Topi: \'old\' glm aic is " + str(round(glm_bin_topi_lag.aic,2))
      + " and lagging glm aic is "+ str(round(glm_bin_topi.aic,2))
     +"\n increased about "+
     "{:.2%}".format(1-(glm_bin_topi_lag.aic/glm_bin_topi.aic)))
print("\nZebra: \'old\' glm aic is " + str(round(glm_bin_Zebra_lag.aic,2))
      + " and lagging glm aic is "+ str(round(glm_bin_zebra.aic,2))+
      "\n increased about "+
     "{:.2%}".format(1-(glm_bin_Zebra_lag.aic/glm_bin_zebra.aic)))
print("\nGazelle: \'old\' glm aic is " + str(round(glm_bin_Gazelle_lag.aic,2))
      + " and lagging glm aic is "+ str(round(glm_bin_gazelle.aic,2))+
      " \n increased about "+
     "{:.2%}".format(1-(glm_bin_Gazelle_lag.aic/glm_bin_gazelle.aic)))
```

AIC for each lagging models are larger than the old models. However, Topi lagging model only increased about 1.48% which is very small. Therefore we have some evidence to conclude that a species were present at a location it may be more likely to be present at the same location 8 days later.

### Is siteID a good variable in the model? <a class="anchor" id="chapter6"></a>

So far we have built our model and checked potential correlation among response variables. Now we turn our focus to the explain variables. $siteID\_cat$ is always significant in each model. However, should we include it in our explain variables? How can we interpret the results in the glm summaries? Will there be any collinearity between  $siteID\_cat$ and other variables?

To answer these questions, we can rebuild our glm models without $siteID\_cat$ and compare AICs with models that have  $siteID\_cat$.

## Topi

```python
X_topi_noSite = Topi[['ndvi','fire','amRivDist','TM100',
                    'LriskDry','T50','month']]
Y_topi_noSite = Topi['present']
glm_bin_topi_noSite = best_subset_glm(X_topi_noSite,Y_topi_noSite)
print(glm_bin_topi_noSite.summary())
```

The above glm model summary shows that amRivDist and LriskDry are significant. Therefore, our full glm model systematic component for Topi presence are

$$
\begin{aligned}

### Summary  <a class="anchor" id="chapter7"></a>

$\textit{What are the primary factors which determine the presence/absence of each of the 3 species?}$

Topi

$amRivDist$ and $LriskDry$ are the primary factors that determine the presence/absence of topi. Specifically, with all other things unchanged, every unit increase in $amRivDist$ will decrease the odds of topi presence by $e^{-0.0003}=0.9997$ times, and every 0.001 unit(since $LriskDry$ value is at 0.001 level) increase in $LriskDry$ will increase the odds of topi presence by $e^{1.61}=5.05$ times.

$$
\begin{aligned}
log\frac{\hat{\mu}_{topi}}{1-\hat{\mu}_{topi}} & = \hat{\beta_0}+\hat{\beta_1}*x_{amRivDist} + \hat{\beta_2}*x_{LriskDry}\\
& =  -2.5573-0.0003*x_{amRivDist}  +161.9892 *x_{LriskDry}
\end{aligned}
$$

Zebra

$TM100$, $T50$ and $month$ are the primary factors that determine the presence/absence of zebra. Specifically, with all other things unchanged, every unit increase in $TM100$ will decrease the odds of topi presence by $e^{-0.2718}=0.762$ times, and every unit increase in $T50$ will increase the odds of topi presence by $e^{0.001}=1.001$ times, and every unit increase in $month$ will increase the odds of topi presence by $e^{0.0505}=1.05$ times. One thing to notice here is that Zebra will show up more as in later of a year and that's because Tanzania is in the south part of the earth.

$$
\begin{aligned}
log\frac{\hat{\mu}_{zebra}}{1-\hat{\mu}_{zebra}} & = \hat{\beta_0}+\hat{\beta_1}*x_{TM100} + \hat{\beta_2}*x_{T50} +  \hat{\beta_3}*x_{month}\\
& =  -0.8747-0.2718*x_{TM100}  +0.0010 *x_{T50}+ 0.0505x_{month}
\end{aligned}
$$

Gazelle

$ndvi$, $amRivDist$ and $TM100$ are the primary factors that determine the presence/absence of gazelle. Specifically, with all other things unchanged, every unit increase in $ndvi$ will decrease the odds of gazelle presence by $e^{-5.5086}=0.004$ times, and every unit increase in $amRivDist$ will increase the odds of gazelle presence by $e^{0.0002}=1.0002$ times, and every unit increase in $TM100$ will increase the odds of gazelle presence by $e^{-0.7992}=0.449$ times. Notice that for gazelle the glm doesn't have an intercept.

$$
\begin{aligned}
log\frac{\hat{\mu}_{gazelle}}{1-\hat{\mu}_{gazelle}} & = \hat{\beta_1}*x_{ndvi} +\hat{\beta_2}*x_{amRivDist} + \hat{\beta_3}*x_{TM100}\\
& = -5.5086 *x_{ndvi} +0.0002*x_{amRivDist}  -0.7992*x_{TM100}
\end{aligned}
$$
